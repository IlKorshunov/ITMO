# Выкладка про λ = 2.35

---

## 1. Функция ценности

$$v(x) = \begin{cases} x^\alpha & x \geq 0 \\ -\lambda(-x)^\beta & x < 0 \end{cases}, \quad \alpha = \beta = 0.88$$

---

## 2. Условие безразличия

$P(+G) = P(-L) = 0.5$. Безразличие при $\mathbb{E}[v] = 0$:

$$\frac{1}{2}G^\alpha = \frac{1}{2}\lambda L^\beta \implies \lambda = \left(\frac{G}{L}\right)^\alpha$$

При $G/L \approx 2.4$: $\quad\lambda = 2.4^{0.88} \approx 2.35$

---

## 3. Отношение градиентов

$$v'(x) = \begin{cases} \alpha x^{\alpha-1} & x > 0 \\ \lambda\beta(-x)^{\beta-1} & x < 0 \end{cases}$$

$$\frac{|v'(-\varepsilon)|}{|v'(+\varepsilon)|} = \frac{\lambda\beta}{\alpha} = \lambda \quad (\alpha = \beta)$$

---

## 4. Модификация Q-learning

$$V(s,a) \leftarrow V(s,a) + \eta\left[v(r - r_0) + \gamma\max_{a'}V(s',a') - V(s,a)\right]$$

$$\frac{\partial V}{\partial r}\bigg|_{r<r_0} = \lambda \cdot \frac{\partial V}{\partial r}\bigg|_{r>r_0}$$
